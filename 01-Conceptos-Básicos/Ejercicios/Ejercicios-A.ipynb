```python
#Ejercicio 2A
##Escribir funciones para tranformar grados Celsius en Fahrenheit y viciversa.

##Formulas: Celsius = (5 ÷ 9) x (Fahrenheit - 32) Fahrenheit = (Celsius x (9 ÷ 5)) + 32


def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

while True:
    try:
        unidad = input("Escriba - C - si desea la temperatura en Celcius \nEscriba - F - Si desea la temperatura en Fahrenheit \n escriba -fin- para acabar:  ")

        unidad = unidad.lower()
        if(unidad == "c"):
            temperaturaC = int(input("Escriba el numero de la temperatua en Fahrenheit para retornar Celcius : "))
            print(f"Los {temperaturaC} ºF son {fahrenheit_to_celsius(temperaturaC):0.2f} ºC")

        elif(unidad == "f"):
            temperaturaF = int(input("Escriba el numero de la temperatua en Celcius para retornar Fahrenheit : "))
            print(f"Los {temperaturaF} ºC son {celsius_to_fahrenheit(temperaturaF):0.2f} ºF")
        
        elif(unidad == "fin"):
            break        

        print("Nuevo calculo.....")        
    except:
        print("Dato introducido incorrectamente")
else:
    print("Fin de la calculadora")
```

```python
#Ejercicio 3A
##Escribe una función que calcule los impuesto a pagar depediendo el salario.

###Tabla: Menos de 10.000 5% Entre 10.000 y 20.000 15% Entre 20.000 y 35.000 20% Entre 35.000 y 60.000 30% Más de 60.000 45%

        #En concepto de desempleo     4%
        #En concepto de formación   1,2%

salarioinput = input("Introducir sueldo de un año")

try:
    salario = float(salarioinput)
except:
    print("Error: datos incorrectos.\nIntroducir numeros")

impuestos = 0
# calculo
if(salario > 0 and salario < 10000):
    impuestos = salario * 0.05
elif(salario < 20000):
    impuestos = salario * 0.15
elif(salario < 35000):
    impuestos = salario * 0.20
elif(salario < 60000):
    impuestos = salario * 0.30
else:
    impuestos = salario * 0.45

print(f"Debes de pagar {impuestos} € en impuestos")
```

```python
#Ejercicio 4A
##Escribe una función que cuente las vocales de una frase.
frase = input("Escribir frase para contar las vocales :  ")

vocales = {
    "a": 0,
    "e": 0,
    "i": 0,
    "o": 0,
    "u": 0
}

print("Desglosando vocales...")
print(frase)
try:
    for ii in vocales:
        vocales[ii] = frase.count(ii)
        print(f"{ii}  :  {vocales[ii]}")

except:
    print("Dato erroneo")
```

```python
import random

menor = 1
mayor = 20
numero1 = random.randint(menor, mayor)

numIntentos = 0
while True:
    try:
        entrada = input(f"Dime un número entre {menor} y {mayor}")
        intento = int(entrada)

        numIntentos += 1
        if(intento > numero1):
            print("Te has pasado")
        elif(intento < numero1):
            print("Te quedaste corto")
        elif(intento == numero1):
            print(f"Acertaste! Has necesitado {numIntentos} intentos")
            break
    except:
        print("Error en la entrada")
```

```python
ingredientes = { "Base": 8.90, "Pepperoni": 0.50, "Pollo": 0.35, "Carne": 0.35, "Pimiento": 0.25, "Cebolla": 0.25, "Champiñon": 0.25, "Aceituna": 0.50, "Piña": 0.35, "ExtraQueso": 1 }

def Pizza():
    pizza = [
        ["base", 8.90, 1]
    ]

    while(True):
        print("\n¿Qué pizza quieres comer?")
        for ii in ingredientes:
            print(ii)
        
        try:
            añade_ingrediente = input("Añade un ingrediente: ")
            if(not añade_ingrediente in ingredientes):
                raise Exception("Ingrediente no disponible, intente con otro")
 
            cantidad = input("Añade cantidad :")
            if(not cantidad.isdigit() ):
                raise Exception("Cantidad debe de ser un número entero")

            stop = ["FIN", "1", "True", "Si"]
            ## Añadie ingredientes:
            pizza.append([añade_ingrediente, ingredientes[añade_ingrediente], int(cantidad)])

            detener = input("¿Parar de confeccionar pizza? (No) :")
            if(detener in stop):
                break

        except Exception as err:
            print(err)
        
        
    ## Calcular valor de la pizza
    print(pizza)

    valor = []
    for ii in pizza:
        valor.append(ii[1] * ii[2])

    return sum(valor)


# _____ MAIN _____
print("~~~ Bienvendo a la Pizzeria ~~~\n")
valor_pizza = Pizza()

print(f"La pizza vale {valor_pizza}")
```
